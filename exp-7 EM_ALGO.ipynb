```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv(r'Downloads\em.csv')
dfl=pd.DataFrame(data)
```

```python
print(dfl)
```

```python
f1 = dfl['sepal.length'].values
f2 = dfl['petal.length'].values
```

```python
X=np.asarray(list (zip (f1, f2)))
```

```python
plt.plot()
plt.xlim([0, 100])
plt.ylim([0, 50])
plt.title('Dataset')
plt.ylabel('sepal.length')
plt.xlabel('petal.length')
plt.scatter(f1,f2)
colors = ['b', 'g', 'r']
markers = ['o', 'v', 's']
```

```python
kmeans_model = KMeans(n_clusters=3). fit (X)
```

```python
plt.plot()
for i, l in enumerate(kmeans_model.labels_):
 plt.plot(f1[i], f2[i], color=colors[l], marker=markers[l],ls='None')
 plt.xlim([0, 10])
 plt.ylim([0, 10])
plt.show()
```
